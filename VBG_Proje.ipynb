```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# üîπ Sƒ±caklƒ±ƒüa g√∂re sezon belirleme
def get_season_from_temperature(temp):
    if temp < 10:
        return "kƒ±≈ülƒ±k"
    elif 10 <= temp < 20:
        return "baharlƒ±k"
    elif 20 <= temp < 28:
        return "yazlƒ±k"
    else:
        return "a√ßƒ±k yazlƒ±k"

# üîπ Trendyol'dan veri √ßekme
def scrape_trendyol_by_season(season, pages=1):
    headers = {"User-Agent": "Mozilla/5.0"}

    keywords = {
        "kƒ±≈ülƒ±k": "kadƒ±n mont kaban",
        "baharlƒ±k": "tren√ßkot kadƒ±n",
        "yazlƒ±k": "kadƒ±n elbise",
        "a√ßƒ±k yazlƒ±k": "askƒ±lƒ± bluz"
    }

    query = keywords.get(season, "kadƒ±n giyim")
    base_url = "https://www.trendyol.com/sr?q="
    products = []

    for page in range(1, pages + 1):
        url = f"{base_url}{query}&pi={page}"
        print(f"üîç Scraping: {url}")
        response = requests.get(url, headers=headers)
        if response.status_code != 200:
            print(f"‚ùå Sayfa {page} alƒ±namadƒ±.")
            continue

        soup = BeautifulSoup(response.text, "html.parser")
        items = soup.find_all("div", class_="p-card-wrppr")

        for item in items:
            title = item.find("span", class_="prdct-desc-cntnr-name")
            brand = item.find("span", class_="prdct-desc-cntnr-ttl")
            price = item.find("div", class_="prc-box-dscntd")

            if title and brand and price:
                products.append({
                    "brand": brand.text.strip(),
                    "title": title.text.strip(),
                    "price": price.text.strip(),
                    "season": season
                })

        time.sleep(1)

    df = pd.DataFrame(products)
    if not df.empty:
        df.to_csv(f"{season}_kadin_kombinleri.csv", index=False, encoding="utf-8-sig")
        print(f"‚úÖ {season} kombinleri CSV olarak kaydedildi.")
    else:
        print("‚ö†Ô∏è Hi√ß √ºr√ºn bulunamadƒ±.")

    return df

# üîπ Hava durumu API'sinden sƒ±caklƒ±k √ßekme
def get_weather_and_run(api_key, city="Eski≈üehir"):
    weather_url = f"http://api.weatherapi.com/v1/current.json?key={api_key}&q={city}&aqi=no"
    response = requests.get(weather_url)

    if response.status_code == 200:
        data = response.json()
        current = data["current"]
        temp = current["temp_c"]

        # Sezon belirle
        season = get_season_from_temperature(temp)
        print(f"\nüß• √ñnerilen sezon: {season} - Sƒ±caklƒ±k: {temp}¬∞C\n")

        # Trendyol‚Äôdan uygun kƒ±yafetleri √ßek
        df = scrape_trendyol_by_season(season)
        if not df.empty:
            print("\nüõçÔ∏è √ñnerilen √úr√ºnler:\n")
            print(df.head())
        else:
            print("√úr√ºn listesi bo≈ü.")
    else:
        print("‚ùå Hava durumu verisi alƒ±namadƒ±.")

# üî∏ API key gir
api_key = "YOUR_WEATHER_API_KEY"  # Buraya kendi weatherapi.com API anahtarƒ±nƒ± yaz
get_weather_and_run(api_key, city="Eski≈üehir")
```

```python
import pandas as pd

API_KEY = "16ce454570964d39901204010250905"  # Buraya kendi WeatherAPI key‚Äôini yaz.
BASE_URL = "http://api.weatherapi.com/v1/current.json"

def fetch_weather(city):
    """Belirtilen ≈üehir i√ßin anlƒ±k hava durumu verisini d√∂nd√ºr√ºr."""
    params = {
        "key": API_KEY,
        "q": city,
        "aqi": "no"
    }
    response = requests.get(BASE_URL, params=params)

    if response.status_code != 200:
        print(f"Hata olu≈ütu: {city} i√ßin veri alƒ±namadƒ±. {response.text}")
        return None

    data = response.json()
    return {
        "city": city,
        "time": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
        "condition": data["current"]["condition"]["text"],
        "temp_c": data["current"]["temp_c"],
        "feelslike_c": data["current"]["feelslike_c"],
        "humidity": data["current"]["humidity"],
        "wind_kph": data["current"]["wind_kph"]
    }

def save_to_csv(df, filename="hava_durumu_verisi.csv"):
    df.to_csv(filename, index=False, encoding="utf-8-sig")
    print(f"{len(df)} ≈üehirden hava durumu verisi '{filename}' dosyasƒ±na kaydedildi.")

selected_city = input("Hava durumu verisi almak istediƒüiniz ≈üehri girin: ")

weather_data = fetch_weather(selected_city)

# DataFrame olu≈ütur
if weather_data:
    df_weather = pd.DataFrame([weather_data]) # fetch_weather tek bir s√∂zl√ºk d√∂nd√ºrd√ºƒü√º i√ßin listeye alƒ±nƒ±r
    print("\n√áekilen Hava Durumu Verisi:")
    print(df_weather)
else:
    print(f"{selected_city} i√ßin hava durumu verisi alƒ±namadƒ±.")
```

```python
def map_temperature_to_season(temp_c):
    """Sƒ±caklƒ±ƒüa g√∂re mevsim etiketi d√∂ner (giyim a√ßƒ±sƒ±ndan)."""
    if temp_c >= 25:
        return "Summer"
    elif 18 <= temp_c < 25:
        return "Spring"
    elif 10 <= temp_c < 18:
        return "Autumn"
    else:
        return "Winter"

def recommend_clothes(weather_row, df_cloths, gender):
    """
    Hava durumuna g√∂re belirtilen cinsiyet i√ßin kƒ±yafet √∂nerisi d√∂ner.
    """
    temp_c = weather_row["temp_c"]
    city = weather_row["city"]
    season = map_temperature_to_season(temp_c)

    filtered = df_cloths[
        ((df_cloths["season"] == season) | (df_cloths["season"] == "All")) &
        (df_cloths["gender"] == gender)
    ]

    if filtered.empty:
        return f"{gender} i√ßin {city} ≈üehrine uygun kƒ±yafet bulunamadƒ±."

    recommended = filtered["category"].unique().tolist()
    return recommended

# ≈ûehir se√ßimi
city_input = input("L√ºtfen bir ≈üehir girin (√∂rnek: Istanbul): ")

# ≈ûehre ait hava durumu verisini bul
city_weather = df_weather[df_weather["city"].str.lower() == city_input.lower()]

if city_weather.empty:
    print(f"{city_input} i√ßin hava durumu verisi bulunamadƒ±.")
else:
    row = city_weather.iloc[0]
    print(f"\n{row['city']} i√ßin hava durumu: {row['temp_c']}¬∞C, {row['condition']}\n")

    male_rec = recommend_clothes(row, df_cloths, gender="Male")
    female_rec = recommend_clothes(row, df_cloths, gender="Female")

    print("üëï Erkekler i√ßin √∂nerilen kƒ±yafetler:", ", ".join(male_rec))
    print("üëó Kadƒ±nlar i√ßin √∂nerilen kƒ±yafetler:", ", ".join(female_rec))
```
