```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# 🔹 Sıcaklığa göre sezon belirleme
def get_season_from_temperature(temp):
    if temp < 10:
        return "kışlık"
    elif 10 <= temp < 20:
        return "baharlık"
    elif 20 <= temp < 28:
        return "yazlık"
    else:
        return "açık yazlık"

# 🔹 Trendyol'dan veri çekme
def scrape_trendyol_by_season(season, pages=1):
    headers = {"User-Agent": "Mozilla/5.0"}

    keywords = {
        "kışlık": "kadın mont kaban",
        "baharlık": "trençkot kadın",
        "yazlık": "kadın elbise",
        "açık yazlık": "askılı bluz"
    }

    query = keywords.get(season, "kadın giyim")
    base_url = "https://www.trendyol.com/sr?q="
    products = []

    for page in range(1, pages + 1):
        url = f"{base_url}{query}&pi={page}"
        print(f"🔍 Scraping: {url}")
        response = requests.get(url, headers=headers)
        if response.status_code != 200:
            print(f"❌ Sayfa {page} alınamadı.")
            continue

        soup = BeautifulSoup(response.text, "html.parser")
        items = soup.find_all("div", class_="p-card-wrppr")

        for item in items:
            title = item.find("span", class_="prdct-desc-cntnr-name")
            brand = item.find("span", class_="prdct-desc-cntnr-ttl")
            price = item.find("div", class_="prc-box-dscntd")

            if title and brand and price:
                products.append({
                    "brand": brand.text.strip(),
                    "title": title.text.strip(),
                    "price": price.text.strip(),
                    "season": season
                })

        time.sleep(1)

    df = pd.DataFrame(products)
    if not df.empty:
        df.to_csv(f"{season}_kadin_kombinleri.csv", index=False, encoding="utf-8-sig")
        print(f"✅ {season} kombinleri CSV olarak kaydedildi.")
    else:
        print("⚠️ Hiç ürün bulunamadı.")

    return df

# 🔹 Hava durumu API'sinden sıcaklık çekme
def get_weather_and_run(api_key, city="Eskişehir"):
    weather_url = f"http://api.weatherapi.com/v1/current.json?key={api_key}&q={city}&aqi=no"
    response = requests.get(weather_url)

    if response.status_code == 200:
        data = response.json()
        current = data["current"]
        temp = current["temp_c"]

        # Sezon belirle
        season = get_season_from_temperature(temp)
        print(f"\n🧥 Önerilen sezon: {season} - Sıcaklık: {temp}°C\n")

        # Trendyol’dan uygun kıyafetleri çek
        df = scrape_trendyol_by_season(season)
        if not df.empty:
            print("\n🛍️ Önerilen Ürünler:\n")
            print(df.head())
        else:
            print("Ürün listesi boş.")
    else:
        print("❌ Hava durumu verisi alınamadı.")

# 🔸 API key gir
api_key = "YOUR_WEATHER_API_KEY"  # Buraya kendi weatherapi.com API anahtarını yaz
get_weather_and_run(api_key, city="Eskişehir")
```

```python
import pandas as pd

API_KEY = "16ce454570964d39901204010250905"  # Buraya kendi WeatherAPI key’ini yaz.
BASE_URL = "http://api.weatherapi.com/v1/current.json"

def fetch_weather(city):
    """Belirtilen şehir için anlık hava durumu verisini döndürür."""
    params = {
        "key": API_KEY,
        "q": city,
        "aqi": "no"
    }
    response = requests.get(BASE_URL, params=params)

    if response.status_code != 200:
        print(f"Hata oluştu: {city} için veri alınamadı. {response.text}")
        return None

    data = response.json()
    return {
        "city": city,
        "time": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
        "condition": data["current"]["condition"]["text"],
        "temp_c": data["current"]["temp_c"],
        "feelslike_c": data["current"]["feelslike_c"],
        "humidity": data["current"]["humidity"],
        "wind_kph": data["current"]["wind_kph"]
    }

def save_to_csv(df, filename="hava_durumu_verisi.csv"):
    df.to_csv(filename, index=False, encoding="utf-8-sig")
    print(f"{len(df)} şehirden hava durumu verisi '{filename}' dosyasına kaydedildi.")

selected_city = input("Hava durumu verisi almak istediğiniz şehri girin: ")

weather_data = fetch_weather(selected_city)

# DataFrame oluştur
if weather_data:
    df_weather = pd.DataFrame([weather_data]) # fetch_weather tek bir sözlük döndürdüğü için listeye alınır
    print("\nÇekilen Hava Durumu Verisi:")
    print(df_weather)
else:
    print(f"{selected_city} için hava durumu verisi alınamadı.")
```

```python
def map_temperature_to_season(temp_c):
    """Sıcaklığa göre mevsim etiketi döner (giyim açısından)."""
    if temp_c >= 25:
        return "Summer"
    elif 18 <= temp_c < 25:
        return "Spring"
    elif 10 <= temp_c < 18:
        return "Autumn"
    else:
        return "Winter"

def recommend_clothes(weather_row, df_cloths, gender):
    """
    Hava durumuna göre belirtilen cinsiyet için kıyafet önerisi döner.
    """
    temp_c = weather_row["temp_c"]
    city = weather_row["city"]
    season = map_temperature_to_season(temp_c)

    filtered = df_cloths[
        ((df_cloths["season"] == season) | (df_cloths["season"] == "All")) &
        (df_cloths["gender"] == gender)
    ]

    if filtered.empty:
        return f"{gender} için {city} şehrine uygun kıyafet bulunamadı."

    recommended = filtered["category"].unique().tolist()
    return recommended

# Şehir seçimi
city_input = input("Lütfen bir şehir girin (örnek: Istanbul): ")

# Şehre ait hava durumu verisini bul
city_weather = df_weather[df_weather["city"].str.lower() == city_input.lower()]

if city_weather.empty:
    print(f"{city_input} için hava durumu verisi bulunamadı.")
else:
    row = city_weather.iloc[0]
    print(f"\n{row['city']} için hava durumu: {row['temp_c']}°C, {row['condition']}\n")

    male_rec = recommend_clothes(row, df_cloths, gender="Male")
    female_rec = recommend_clothes(row, df_cloths, gender="Female")

    print("👕 Erkekler için önerilen kıyafetler:", ", ".join(male_rec))
    print("👗 Kadınlar için önerilen kıyafetler:", ", ".join(female_rec))
```
